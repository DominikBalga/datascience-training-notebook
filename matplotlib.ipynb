```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.title("title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
# OO matplot
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("y Label")
axes.set_title("Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.1,0.4,0.3])
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
# yaxes.plot(x,y)
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title("first plot")
axes[1].set_title("second plot")
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(8,4))
axes[0][0].plot(x,y)
plt.tight_layout()
```

```python
fig.savefig("my_pic.png")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label="X with Y")
# ax.set_title("Random")
# ax.set_xlabel("random x")
ax.plot(y,x, label = "Y with X",linewidth=4)
# ax.set_ylabel("random y")
ax.legend(loc=0)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_xlim([0,1])
axes.set_ylim([0,1])
axes.plot(x,y,color="#FF8C00", lw=3,alpha=0.5,ls="dashed",marker="o",ms=10,mfc="red",mec="black")
```
